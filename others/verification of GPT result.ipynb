```python
features = [ 'Biomass Generation (MW)', 'Coal Generation (MW)', 'Gas Generation (MW)', 'Gas-CC Generation (MW)', 'Hydro Generation (MW)', 'Solar Generation (MW)', 'Bus average LMP_y', 'Houston LMP_y', 'Hub average LMP_y', 'North LMP_y', 'Panhandle LMP_y', 'South LMP_y', 'West LMP_y', 'Average Temperature (Fahrenheit)', 'Hour', 'Month_sin', 'Month_cos', 'Hour_sin']

lag_hours = 24

# Create lagged features for each of the specified features
lagged_features = {}
for feature in features:
    lagged_features[feature + '_lag1'] = df[feature].shift(lag_hours)

# Add the lagged features to the dataframe
df_lagged = df.copy()
df_lagged = df_lagged.assign(**lagged_features)
```

```python
df['Local Date'] = pd.to_datetime(df['Local Timestamp Central Time (Interval Beginning)'], format='%m/%d/%Y %H:%M')

df['Year'] = df['Local Date'].dt.year

yearly_load = df.groupby('Year')['Total Generation (MW)'].sum()

for year in [2021, 2022, 2023, 2024]:
    if year in yearly_load:
        print(f"Total Generation for {year}: {yearly_load[year]} MW")
    else:
        print(f"No data available for the year {year}")

import pandas as pd
```
