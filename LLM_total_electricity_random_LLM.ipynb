```python
# Tokenize the inputs and labels (explanations) with padding and truncation
def preprocess_function(examples):
    inputs = [str(ex) for ex in examples['Predicted Energy Load']]
    targets = [str(ex) for ex in examples['Explanation']]

    # Tokenize inputs
    model_inputs = tokenizer(inputs, max_length=512, padding='max_length', truncation=True)

    # Tokenize targets (explanations) and pad to the same length
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=512, padding='max_length', truncation=True)

    model_inputs['labels'] = labels['input_ids']
    return model_inputs

# Apply the preprocessing to the dataset
tokenized_dataset = hf_dataset.map(preprocess_function, batched=True)

```
